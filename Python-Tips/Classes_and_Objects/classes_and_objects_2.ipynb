# Classes and Objects

**problem**: 

In a hotel, there are guests coming with following registrations:
```
Name - John
Key - A011

Name - Kyle
Key - A009

Name - Kyle
Key - BQ02

Name - Kyle
Key - A015

Name - Kyle
Key - BQ03
```

- Create a `Guest` class and store the above values in it. While initializing pass the guest name and store it as `name`.
- Create a method `is_regd()` to print `'Registered'` or `'Not Registered'` when the guest is registered or not respectively.
- Create a method `get_key()` to print the key of the guest and print `Not Registered` if not.
- Create another method, `reg()` to register the guest and assign a key to that guest from the below list of Keys

```python
kys = ['A010', 'A012', 'A014', 'BQ01']
```

Also, print a message `'Sorry, no vacant rooms available` if the `kys` list becomes empty.

Then create `Guest`'s for the following guests:
* Josh
* Hans
* Evan
* Kyle
* Ted
* Karl
* Sam

And print the name of the guest with it's registration condition, print the Key if registered else register and then print the key like:

```
Kyle
Registered
Key : A009
```


````python
class Guest:
    
    regs= {'John' : 'A011',
           'Kyle' : "A009",
           'Jake' : 'BQ02',
           'Tamra': 'A015',
           'Josh' : 'BQ03'
          }
    
    kys = ['A010', 'A012', 'A014', 'BQ01']
    
    def __init__(self, name):
        self.name = (name.lower()).capitalize()
        self.registered = self.name in self.regs
        
    def is_regd(self):
        if self.registered:
            print("Registered")
        else:
            print("Not Registered")
            
    def get_key(self):
        if self.registered:
            print(f'Key : {self.regs[self.name]}')
            
    def reg(self):
        if len(self.kys):
            self.regs[self.name] = self.kys[0]
            self.kys.pop(0)
            self.registered = True
        else:
            print("Sorry, no vacant rooms available")
    
````

````python
names = ["Josh", "Hans", "Evan", "Kyle", "Ted", "Karl", "Sam"]

for name in names:
    guest = Guest(name)
    print(guest.name)
    guest.is_regd()
    
    if guest.registered:
        guest.get_key()
    else:
        guest.reg()
        guest.get_key()
````
